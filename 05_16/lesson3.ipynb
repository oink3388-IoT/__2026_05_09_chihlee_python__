```python
#學生總分為300分
#有些學分可以加分5%


while True:
    try:
        scores=int(input("請輸入學生分數(最高300分):"))
        break
    except ValueError:
        print("輸入錯誤，請輸整數！")

is_add = input("學生是否符合加分條件？(y,n)")

if is_add == "y": #單向選擇
    score = scores * 1.05

#四捨五入
scores = int(score+0.5)

#超過300分，上限取300分
if scores > 300:
    scores = 300   

print("學生分數:",scores)



```
