```python
import tkinter as tk
from tkinter import messagebox
import random

secret = random.randint(1, 100)
count = 0

def check_guess(event=None):
    global count, secret

    text = entry.get().strip()

    if not text:
        result_label.config(text="請先輸入數字", fg="red")
        return

    try:
        guess = int(text)
    except ValueError:
        result_label.config(text="請輸入有效整數", fg="red")
        return

    count += 1

    if guess < secret:
        result_label.config(text=f"第 {count} 次：太小了", fg="#1f77b4")
    elif guess > secret:
        result_label.config(text=f"第 {count} 次：太大了", fg="#d62728")
    else:
        result_label.config(text=f"第 {count} 次：恭喜猜對了！答案是 {secret}", fg="#2ca02c")
        guess_button.config(state="disabled")

def reset_game():
    global secret, count
    secret = random.randint(1, 100)
    count = 0
    entry.delete(0, tk.END)
    result_label.config(text="請輸入 1 到 100 的數字開始猜", fg="#333333")
    guess_button.config(state="normal")

root = tk.Tk()
root.title("猜數字遊戲")
root.geometry("420x260")
root.resizable(False, False)
root.configure(bg="#f5f7fa")

title = tk.Label(root, text="猜數字遊戲", font=("Microsoft JhengHei", 20, "bold"), bg="#f5f7fa", fg="#222222")
title.pack(pady=(18, 8))

hint = tk.Label(root, text="請輸入 1 到 100 的整數", font=("Microsoft JhengHei", 11), bg="#f5f7fa", fg="#555555")
hint.pack()

entry_frame = tk.Frame(root, bg="#f5f7fa")
entry_frame.pack(pady=12)

entry = tk.Entry(entry_frame, font=("Microsoft JhengHei", 14), width=14, justify="center", bd=2, relief="solid")
entry.pack(side="left", padx=8)
entry.bind("<Return>", check_guess)

guess_button = tk.Button(
    entry_frame,
    text="送出",
    font=("Microsoft JhengHei", 12),
    bg="#4c8bf5",
    fg="white",
    activebackground="#3b76d1",
    activeforeground="white",
    padx=14,
    pady=4,
    command=check_guess
)
guess_button.pack(side="left")

result_label = tk.Label(
    root,
    text="請輸入 1 到 100 的數字開始猜",
    font=("Microsoft JhengHei", 12),
    bg="#f5f7fa",
    fg="#333333"
)
result_label.pack(pady=15)

reset_button = tk.Button(
    root,
    text="重新開始",
    font=("Microsoft JhengHei", 11),
    bg="#6c757d",
    fg="white",
    activebackground="#5a6268",
    activeforeground="white",
    padx=12,
    pady=4,
    command=reset_game
)
reset_button.pack()

entry.focus()
root.mainloop()
```
