```python
import random

secret = random.randint(1, 100)
count = 0

print("歡迎來玩猜數字遊戲！")
print("我已經想好 1 到 100 之間的一個數字了。")

while True:
    guess = int(input("請輸入你猜的數字："))
    count += 1

    if guess < secret:
        print("太小了，再試一次。")
    elif guess > secret:
        print("太大了，再試一次。")
    else:
        print(f"恭喜你猜對了！答案是 {secret}。")
        print(f"你總共猜了 {count} 次。")
        break
```
