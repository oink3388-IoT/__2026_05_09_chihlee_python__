```python
#建立工廠

class Persons:
    #pass
    #實體方法
    def __init__(self,name): #自訂初始化
        print("初始化：一個新的人被建立了") 
        self.name = name
    #自訂實體的method
    def introduce(self):
        print(f"Hello, 我的名字是{self.name}")
        
```

```python
p4=Persons("Steve")
p4.introduce()
```
