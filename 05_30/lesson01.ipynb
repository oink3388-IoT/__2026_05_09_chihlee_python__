```python
#資料結構
#程式結構

#List
#可以修改
#也index從0開始

scores=[89,92,76,91,71]
print(scores)
print(scores[0])
print(scores[1])
print(scores[2])
print(scores[3])
print(scores[4])

scores[0]=100

#for迴圈
for i in range(5):
    print(scores[i])    
#for迴圈簡化
for score in scores:
    print(score) 
#資料結構：串列、字典、集合、元組
#串列：有順序、可重複、可變動
scores=[89,92,76,91,71]
print(scores)   
scores.append(85)
print(scores)
scores[0]=90
print(scores)   
#字典：無順序、可重複、可變動
student={"name":"Alice","age":20,"grade":"A"}
print(student)


```
