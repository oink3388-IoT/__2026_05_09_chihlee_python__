```python
#高中數學
'''
Homerwork(degree.py)
讓使用者輸入直角三角形的對邊
讓使用者輸入直角三角形的斜邊
計算角度
公式
sin(x)=三角形對邊/斜邊
x是radian(弧度)

'''

import math

opposite=float(input("請輸入直角三角形的對邊: "))
hypotenuse=float(input("請輸入直角三角形的斜邊: "))

sin_x=opposite/hypotenuse
x_radian=math.asin(sin_x)

#使用round() 進行四捨五入
print("三角形的弧度是：",round(x_radian,2))
print("三角形的角度是：",round(math.degrees(x_radian),2))
```
