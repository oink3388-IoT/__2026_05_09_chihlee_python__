```python
import numpy as np
from numpy import ndarray
import pandas as pd
from pandas import DataFrame

scores:ndarray | None = None #學生的分數
scores_df:DataFrame | None = None #學生的分數的DataFrame
np.random.seed(42)  # 加在 randint 之前
scores=np.random.randint(50,101, size=(50,5))
#scores
#print(type(scores))
#print(scores.shape)
scores_df=pd.DataFrame(scores,
         columns=['國文','英文','數學','地理','歷史'],
         index=range(1,51))
#scores_df.T
names_df:DataFrame=pd.read_csv('students.csv')
#type(names_df['姓名'])
scores_df[['姓名','性別']]=names_df[['姓名','性別']].head(n=50).values
scores_df=scores_df[['姓名','性別','國文','英文','數學','地理','歷史']]
#選取欄
scores_df[['姓名','國文']]
#選取列
scores_df.iloc[5]
#slice寫法
scores_df.iloc[5:10]
#scores_df[scores_df['姓名']=='鄧文龍']
#scores_df[scores_df['國文']<60]
#display(scores_df.query("性別=='男' and 國文<60"))
test_df=scores_df.query("性別=='女' and 國文<60")
test_df=scores_df[['國文','英文','數學','地理','歷史']]
scores_df["總分"]=scores_df[['國文','英文','數學','地理','歷史']].sum(axis=1)
scores_df["平均"]=scores_df[['國文','英文','數學','地理','歷史']].mean(axis=1)
#display(scores_df.query("平均>80"))
scores_df["排名"]=scores_df["總分"].rank(ascending=False,method='min')
scores_df.sort_values(by='排名')


```

```python
from pandas import Series
def calculator(s:Series) -> int :
    n=((s<60).sum())
    return n

scores_df['不及格的科目數']=scores_df[['國文','英文','數學','地理','歷史']].apply(calculator,axis=1)
scores_df
```

```python
def get_failing_subjects(row):
    subjects=['國文','英文','數學','地理','歷史']
    failing=[subj for subj in subjects if row[subj] < 60] #list綜合表達式
    fallings=['國文','英文','數學']
    return ','.join(failing) if failing else "無"

scores_df["不及格的科目名稱"]=scores_df[['國文','英文','數學','地理','歷史']].apply(get_failing_subjects,axis=1)
scores_df

```
