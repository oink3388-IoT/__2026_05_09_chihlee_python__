```python
import numpy as np
import pandas as pd

# 顯示設定（中文字對齊）
pd.set_option('display.unicode.ambiguous_as_wide', True)
pd.set_option('display.unicode.east_asian_width', True)
pd.set_option('display.float_format', '{:.2f}'.format)

# 產生隨機分數
scores = np.random.randint(50, 101, size=(50, 5))
score_df = pd.DataFrame(scores,
         columns=['國文','英文','數學','地理','歷史'],
         index=range(1, 51))

# 轉成長格式
long_df = score_df.reset_index().melt(
    id_vars='index',
    var_name='科目',
    value_name='分數'
)
long_df.rename(columns={'index': '學生'}, inplace=True)

# 新增班級
np.random.seed(42)
long_df['班級'] = np.random.choice(['A班','B班','C班'], size=len(long_df))

# 樞紐分析表
pivot_table = long_df.pivot_table(
    values='分數',
    index='班級',
    columns='科目',
    aggfunc='mean'
).round(2)

print("=== 原始長格式資料 ===")
print(long_df.head(10))

print("\n=== 樞紐分析表（各班各科平均分）===")
print(pivot_table)
```
