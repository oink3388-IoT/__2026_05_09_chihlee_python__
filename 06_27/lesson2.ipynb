```python
import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values=[20,30,45,10]
colors=['yellow','green','red','blue']

figure = plt.figure()
axes = figure.add_subplot(1,1,1)
#axes.pie(values, labels=labels, colors=colors, autopct='%1.1f%%')
axes.pie(values, labels=labels, colors=colors,explode=(0.3,0,0,0),shadow=True, autopct='%2.1f%%',startangle=90)



```
