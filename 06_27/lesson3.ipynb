```python
import pandas as pd

dataFrame = pd.read_csv('各鄉鎮市區人口密度.csv')
dataFrame.columns = ["統計年", "區域別", "年底人口數", "土地面積", "人口密碼"]

# 把年底人口數轉成數字，無法轉換的變成 NaN
dataFrame["年底人口數"] = pd.to_numeric(dataFrame["年底人口數"], errors="coerce")

# 篩選：年底人口數不是空值、不是 0
df = dataFrame[(dataFrame["年底人口數"].notna()) & (dataFrame["年底人口數"] != 0)]

# 去掉統計年欄位
df2 = df.drop(["統計年"], axis=1)

print(df2)
df2.info()
```

```python
dataFrame.columns = ["統計年", "區域別", "年底人口數", "土地面積", "人口密度"] #改變欄位名稱
df = dataFrame.drop(index=0)
# 1. 移除統計年
df2 = df.drop('統計年', axis=1)
df2
# 2. 將「年底人口數」強制轉為數字，像 '…' 這樣的文字會被強制變成 NaN
df2["年底人口數"] = pd.to_numeric(df2["年底人口數"], errors="coerce")
df2
# 3. 這時候再用 dropna()，就能把原本是 '…' 的東沙、南沙群島（以及其他原本就是 NaN 的列）一次清乾淨！
df2 = df2.dropna(subset=["年底人口數"])
df2.info()
# # 4. 最後順利轉換為整數
df2["年底人口數"] = df2["年底人口數"].astype("int64")
df2["土地面積"] = df2["土地面積"].astype("float")
df2["人口密度"] = df2["人口密度"].astype("int64")
df2
```

```python
import pandas as pd
import numpy as np

# 讀取 CSV 檔
dataFrame = pd.read_csv('各班圖書借閱統計.csv')

# 第 0 列(index 為 0)才是真正的中文欄位名稱,把它設為欄位名
dataFrame.columns = dataFrame.loc[0]

# 刪除第 0 列(已經拿來當欄位名,資料中不需要重複)
dataFrame = dataFrame.drop(0)

# 刪除合計列與 4 列說明文字
dataFrame.drop([dataFrame.index[-5], dataFrame.index[-4], dataFrame.index[-3], dataFrame.index[-2], dataFrame.index[-1]], inplace=True)

# 取得目前的列數與欄數
rowCount, columnCount = dataFrame.shape

# 把索引重設為從 0 開始的連續編號
dataFrame.index = range(rowCount)

# 只保留「班級」「學生人數」「借閱冊數」三個欄位
dataFrame1 = dataFrame.reindex(columns=['班級', '學生人數', '借閱冊數'])

print(dataFrame1)
```
