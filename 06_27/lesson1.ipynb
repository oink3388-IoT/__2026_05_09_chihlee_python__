```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 5, 0.1)
y1 = np.sin(t * np.pi)
y2 = np.cos(t * np.pi)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

ax1.plot(t, y1, color='red', linestyle='-')
ax1.set_title('sin(t * pi)')
ax1.grid(True)

ax2.plot(t, y2, color='green', linestyle='-')
ax2.set_title('cos(t * pi)')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['sin', 'cos']
sizes = [40, 60]
colors = ['#ff6b6b', '#4dabf7']
explode = (0.08, 0.08)

fig, ax = plt.subplots(figsize=(3, 3))

wedges, texts, autotexts = ax.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2},
    textprops={'fontsize': 12, 'color': 'black'}
)

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

ax.set_title('Beautified Pie Chart', fontsize=16, fontweight='bold')
ax.axis('equal')

plt.tight_layout()
plt.show()
```
